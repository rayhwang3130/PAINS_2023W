```python
import pandas as pd
import requests
from lxml import html, etree
from tqdm import tqdm
from selenium.webdriver import *
from selenium import webdriver

first_url='https://fbref.com/en/comps/9/2019-2020/schedule/2019-2020-Premier-League-Scores-and-Fixtures'
```

```python
from user_agent import generate_user_agent, generate_navigator

agent=generate_user_agent(os='win', device_type='desktop')
header={'User-Agent':agent}
r = requests.get(first_url,headers=header)
tree=html.fromstring(r.content)
match_links=tree.xpath('//*[@id="sched_2019-2020_9_1"]/tbody/tr/td[6]/a')
match_links=['https://fbref.com'+x.get('href') for x in match_links]
```

```python
pd.DataFrame(match_links).to_csv('url_links.csv',index=False)
```

```python
match_links=pd.read_csv('url_links.csv').values.tolist()
```

```python
import time
from IPython.display import clear_output

df=pd.DataFrame()

for idx, x in tqdm(enumerate(match_links)):

    print('Pausing for five seconds')

    time.sleep(5)

    if idx%50==0 and idx>0:
        print('Pausing for sixty Seconds')
        time.sleep(60) 

    x=x[0]

    try: 
        r = requests.get(x,headers=header)

    except:
        df.to_csv('Records.csv',index=False)
        
    tree=html.fromstring(r.content)
    match_info=tree.xpath('//*[@id="content"]/h1/text()')[0]

    table=pd.DataFrame(pd.read_html(x,attrs={'id':'shots_all'},header=1)[0])
    table=table[table['Outcome'].notna()]


    target_row=table[table['Outcome'].str.contains('Goal')&table['Event'].str.contains('Pass')&table['Event'].str.contains('Live')]
    target_row=target_row[['Minute','Player','Squad','xG','Distance','Body Part','Player.1','Event']]
    target_row.insert(0,'Match-Up','')
    target_row['Match-Up']=match_info

    script=f'''
    
    [ {idx+1} / {len(match_links)} Records Completed ]
    
    '''
    
    print(script)

    df=pd.concat([df,target_row])
    
    print(df)
```

```python
df.to_csv('Records.csv',index=False)
```
