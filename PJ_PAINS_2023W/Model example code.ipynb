# 모델링

```python
# 메타데이터 설정
from autogluon.tabular import FeatureMetadata
feature_metadata = FeatureMetadata.from_df(train_data)

print(feature_metadata)
```

```python
feature_metadata = feature_metadata.add_special_types({image_col: ['image_path']})

print(feature_metadata)
```

```python
from autogluon.tabular.configs.hyperparameter_configs import get_hyperparameter_config
hyperparameters = get_hyperparameter_config('multimodal')

hyperparameters
```

```python
from autogluon.tabular import TabularPredictor
predictor = TabularPredictor(label=label).fit(
    train_data=train_data,
    hyperparameters=hyperparameters,
    feature_metadata=feature_metadata,
    time_limit=900,
)
```

```python
predictor.get_model_best()
```

```python
predictor.leaderboard(test_data, extra_metrics=['root_mean_squared_error'], silent=True)
```

NOTE: Metrics scores always show in higher is better form. This means that metrics such as `log_loss` and `root_mean_squared_error` will have their signs **FLIPPED**, and values will be negative. This is necessary to avoid the user needing to know the metric to understand if higher is better when looking at leaderboard.
-> 우리가 돌린 모델은 "회귀" 모델이므로 RMSE가 음수로 나온것을 양수로 받아들이면 되는듯

```python
test_data['prediction'] = predictor.predict(test_data)
test_data.head(10)
```

```python
total = pd.concat(
   map(pd.read_csv, [kane, kdb, taa]), ignore_index=True)
total['prediction'] = predictor.predict(test_data)
total.drop(columns=['Match-Up', 'Minute', 'Player', 'Squad', 'Event', 'input_dir', 'crop_dir', 'output_dir', 'dist', 'angle', 'player_dist', 'Distance', 'Body Part'], inplace=True)
total.dropna(axis=0, inplace=True)
total.head(10)
```

```python
# 시각화
import matplotlib.pyplot as plt
import seaborn as sns
sns.lmplot(data=total, x='xG', y='prediction', hue='Player.1')
plt.title("xG and Model Prediction")
plt.ylabel('Prediction')
```

## 공을 움직이며 실제 xG값과 모델이 예측한 xG값 계산

```python
# 데이터 불러오기
adj_kane = pd.read_csv('predicted_xG/kane.csv', index_col=0)
adj_kdb = pd.read_csv('predicted_xG/KDB.csv', index_col=0)
adj_taa = pd.read_csv('predicted_xG/arnold.csv', index_col=0)
```

```python
adj_kane.head()
```

#### 케인

```python
kane_final = adj_kane.groupby('Real_xG', as_index=False).agg(max_model = ('Predicted_xG', 'max'))
kane_final
```

```python
kane_final['difference'] = kane_final['Real_xG'] - kane_final['max_model']
kane_score = kane_final['difference'].sum()
kane_score
```

#### 덕배

```python
kdb_final = adj_kdb.groupby('Real_xG', as_index=False).agg(max_model = ('Predicted_xG', 'max'))
kdb_final
```

```python
kdb_final['difference'] = kdb_final['Real_xG'] - kdb_final['max_model']
kdb_score = kdb_final['difference'].sum()
kdb_score
```

#### 아놀드

```python
taa_final = adj_taa.groupby('Real_xG', as_index=False).agg(max_model = ('Predicted_xG', 'max'))
taa_final
```

```python
taa_final['difference'] = taa_final['Real_xG'] - taa_final['max_model']
taa_score = taa_final['difference'].sum()
taa_score
```

### 해석
- 공을 움직여보며 모델에 넣었을 때 도출되는 xG 값 (Predicted_xG)
- 실제 득점 확률값이 각각의 케이스를 의미하는 바이므로 이를 기준으로 그룹화 (groupby)
- 모델 예측치 중 최고치만 추출 -> 이는 공을 움직였을 때 가장 득점 확률이 높은 곳을 도출해주는 효과
- 각 케이스의 실제 득점 확률과 모델 예측치 중 최고값의 차 (diff)를 구하고 이를 합한 값을 도출 (score)  

#### score은 곧 실제 상황에서 득점확률과 공을 다른 곳에 줬을 때의 득점 확률 중 최고치의 차의 합으로, 
#### 클수록 다른 곳보다 그 상황에서 그 사람에게 주는 것이 옳은 판단이었음을 나타냄
#### 즉, score값이 클수록 패스를 잘한 것임
